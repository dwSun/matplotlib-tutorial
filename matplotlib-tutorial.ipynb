## 简介

数据可视化是数据分析的一个重要的组成部分。随着大数据，互联网的发展以及近年物联网的蓬勃发展，各式各样的海量数据涌入各种商业或非商业机构的存储系统，对海量数据的分析也成为了很多机构日常工作的一部分。而作为数据分析的前置步骤，数据可视化可以提供数据的各种表征的直观表示，因此数据可视化在数据分析中扮演一个非常重要的角色。

为什么可视化如此重要？

人类作为一种视觉动物，视觉是最简单而且直接的理解数据的方式。通过可视化，我们可以非常直观的理解复杂的问题，识别特定模式，相关性，异常值等等。同时，通过可视化呈现的数据，以及其中包含的业务逻辑和销售预期，也更加容易被受众理解和接受。也是因为这些原因，数据可视化是许多高阶任务，如数据分析Exploratory Data Analysis(EDA)和机器学习Machine Learning(ML)等任务的前置步骤。

## Matplotlib和Seaborn

matplotlib是一个主要面向2D绘图的python扩展包。熟悉matlab的人员，会对matplotlib的api使用比较熟悉，因为matplotlib对标的就是matlab的绘图功能。但仅仅为了绘图功能而去学习matlib的语法没有必要，同时matlab也不是免费的。 matplotlib提供了与matlab类似的绘图功能，而且简单易用，更重要的是matplotlib是完全免费的。

Seaborn则是一个建立在 Matplotlib基础之上的高级绘图库，基于 matplotlib已经提供的各种 api，seaborn做了一定的封装，对各种图像的api加入了很多方便的参数而且让各种图像的api高度统一，让绘图变得更加的简单直白。同时 seaborn与数据处理常用的 pandas库进行了无缝的集成，例如某些api直接拥有对输入数据进行统计处理的功能，不需要在额外处理数据，更加方便数据的处理和可视化的过程。这一点与 matplotlib不同，matplotlib专注于绘图，数据处理则要调用者根据需要自己进行。

> matplotlib是包括 seaborn在内的许多绘图库的基础，换言之，seaborn之类的绘图库能绘制的图像，matplotlib都能绘制，但是可能会稍微麻烦一点。

### 一个简单的例子
下面用简单的三角函数曲线来展示一下matplotlib的绘图。

````python
# 后面要用到的各种库导入
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
````

````python
x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值


plt.figure(figsize=(12, 8))  # 创建一个图像，其实这里可以不写
# figsize是以inch为单位的，所以大部分时候，都需要根据实际现实效果调整。
plt.plot(x, y1, 'ro-', x, y2) 
# plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)，
# fmt参考matplotlib相关文档：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.title('sin(x,y)')
plt.legend(['sin', 'cos'])
plt.xlabel('x')
plt.annotate(r'$cos(0)=1$', xy=(0, 1), xytext=(0.1, 0.2),
             arrowprops=dict(facecolor='black', shrink=0.1, width=1))
# plt.annotate() 中'$$'表示使用Latex格式，xy为箭头指向位置，xytext为文本显示位置，facecolor为箭头颜色，shrink为箭头缩进比率，width为箭头宽度。
plt.ylabel('y')
plt.yticks([-1, 0, 1], ['min', 'zero', 'max'])
plt.show()  # 在jupyter中图像是可以自动显示的，所以很多时候，这一句可以不写。
# 但是在这个例子里面，plt.show()如果不写的话，上面的很多设置都无法反映到最终的图像上。
````

## matplotlib中的图像布局

figure作为一个图像窗口，里面可以有多个图像，只需要创建多个axes并对其进行合理的布局即可，如下图：

![](multi_axes.png)

下面介绍matplotlib中图像布局规则和api示例。

### fig.add_subplot/网格布局

matplotlib中，图像以横竖网格布局，以subplot为例

fig.add_subplot(nrows, ncols, index)

参数含义如下
- nrows，图像网格有几行
- ncols，图像网格有几列
- index，图像编号索引，index按照先行后列的顺序，从1开始计数。


由nrows，ncols来确定有几个图像在同一个figure上，而index则表示图像的实际位置。需要注意的是，subplot的布局比较简单，可以利用nrows/ncols对图像进行划分，实现简单的布局，举例如下：    


````python
x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值

fig  = plt.figure()
ax = fig.add_subplot(1, 2, 1)  # 这里绘制一个1行，2列图像的第一个图像，编号为1
ax.plot(x, y1, x, y2)
# 这一次绘图，将整个图像划分成了左右两个部分

ax = fig.add_subplot(222)  # 这里绘制的是一个2行，2列图像的第二列，编号为2的图像
# rows/ncols/index也可以用一个3位的整数代替，每一位分别对应nrows/ncols/index
ax.plot(x, y1, x, y2)
# 这一次绘图，将整个图像划分成了4个部分，但是左边2个部分已经被上一步图像绘制填充。

ax = fig.add_subplot(2, 2, 4)  # 这里绘制的是一个2行，2列图像的第二，编号为4的图像
ax.plot(x, y1, x, y2)
# 注意这里，虽然左边图像已经被占据，而且这里只有3个图像，
# 但是index编号规则是与nrows/ncols确定的绘图位置有关的，所以这里是4,而不是3

plt.show()
````

plt.subplot可以实现同样的功能，而且代码更加简洁。

````python
x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值

plt.subplot(1, 2, 1)  # 这里绘制一个1行，2列图像的第一个图像，编号为1
plt.plot(x, y1, x, y2)
# 这一次绘图，将整个图像划分成了左右两个部分

plt.subplot(222)  # 这里绘制的是一个2行，2列图像的第二列，编号为2的图像
# ows/ncols/index也可以用一个3位的整数代替，每一位分别对应nrows/ncols/index
plt.plot(x, y1, x, y2)
# 这一次绘图，将整个图像划分成了4个部分，但是左边2个部分已经被上一步图像绘制填充。

plt.subplot(2, 2, 4)  # 这里绘制的是一个2行，2列图像的第二，编号为4的图像
plt.plot(x, y1, x, y2)
# 注意这里，虽然左边图像已经被占据，而且这里只有3个图像，
# 但是index编号规则是与nrows/ncols确定的绘图位置有关的，所以这里是4,而不是3

plt.show()
````

大部分情况下，这种布局已经足够用，极少数情况需要更复杂的布局，可以使用plt.subplot2grid()

plt.subplot2grid(shape, loc, rowspan=1, colspan=1)

其中参数含义如下：
- shape为tuple(m,n)，表示将网格分为m行n列
- loc为tuple(a,b)，表示选择第几个网格，索引从0开始，注意这里是横纵索引，不是index编号索引
- rowspan和colspan表示跨几个单元格，默认为1

> 太复杂的图像布局方式其实不利于数据展示，这里仅用于举例，实际不要过多使用这种的布局方式。

````python
x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值

plt.figure(figsize=(7,5))

plt.subplot2grid((3, 3), (0, 0), colspan=3)
plt.plot(x, y1, x, y2)
plt.legend(['Axes1'])

plt.subplot2grid((3, 3), (1, 0), colspan=2)
plt.plot(x, y1, x, y2)
plt.legend(['Axes2'])

plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.plot(x, y1, x, y2)
plt.legend(['Axes3'])

plt.subplot2grid((3, 3), (2, 0), rowspan=2)
plt.plot(x, y1, x, y2)
plt.legend(['Axes4'])

plt.subplot2grid((3, 3), (2, 1), rowspan=2)
plt.plot(x, y1, x, y2)
plt.legend(['Axes5'])
plt.show()
````

### 绝对位置布局和add_axes
网格的布局方式可以让图像看上去更加的紧凑，但是很多时候，我们想在某些特定位置绘图，这就需要 add_axes
fig.add_axes(rec)

其中的参数rect为 [left, bottom, width, height] 
含义分别为
- left 0~1,图像左边起始位置
- botton 0~1,图像底部起始位置
- width 0~1,图像宽
- height 0~1,图像高

````python
x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值

fig  = plt.figure()

ax = fig.add_axes([0,0,1,1])  # 在figure里面绘制一个跟figure一样大的图像
ax.plot(x, y1, x, y2)

ax = fig.add_axes([0.6,0.6,0.3,0.3])  # 在figure右上角绘制一个宽高分别为figure宽高30%的子图。
ax.plot(x, y1, x, y2)

plt.show()
````

### 混用网格布局和绝对布局

fig.add_subplot和fig.add_axes甚至包括plt.subplot都是可以混用的

````python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)  # 创建100个点作为x
y1 = np.sin(x)  # 计算sin值
y2 = np.cos(x)  # 计算sin值

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(241)
ax.plot(x, y1, x, y2)

plt.subplot(242)
plt.plot(x, y1, x, y2)

ax = fig.add_subplot(223)
ax.set_title("subplots")
ax.plot(x, y1, x, y2)

ax = fig.add_axes([0.77,.3,.2,.6])
ax.plot(x, y1, x, y2)

ax =fig.add_axes([0.67,.5,.2,.3])
ax.plot(x, y1, x, y2)

ax = fig.add_axes([0.6,.1,.35,.3])
ax.plot(x, y1, x, y2)
ax.set_title("random axes")

plt.tight_layout() # 自动微调布局，以最合适的方式显示图像
plt.show()
````

## LineChart/折线图

用折线图画时间序列数据，基本就是把所有的数据点用线连起来而已。折线图特别适用于查看数据的变化趋势。上升的曲线意味着增长趋势，下降的曲线意味着衰减趋势，同时，折线图也经常被用来绘制模型性能曲线。

例如，股票数据就是比较典型的时间序列数据，这里用一个例子来展示一下折线图的使用。

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

stock_data = pd.read_csv('datasets/stocks.csv')
````

````python
stock_data.head()
````

````python
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
# 为了绘图方便，这里将文本类型的时间标记转换成pandas的datetime类型
stock_data.head()
````

在同一个图里面展示两个股票的数据

````python
fig = plt.figure(figsize=(10, 6))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('AAPL vs IBM(inset)') # 注意这个title是属于ax1的

ax1.plot(stock_data['Date'],
         stock_data['AAPL'],
         color='green')
ax1.legend()  # 这里是两个不同的axes，所以要分别设置图例

ax2.plot(stock_data['Date'],
         stock_data['IBM'],
         color='blue')
ax2.legend()  # 这里是两个不同的axes，所以要分别设置图例

plt.show()
````

这里，fig.add_axes用来添加多个绘图区域，这里（https://stackoverflow.com/questions/43326680/what-are-the-differences-between-add-axes-and-add-subplot） 可以看到add_axes and add_subplot的区别。
plt.plot的第一个参数会被用作横轴，第二个数据会被用作纵轴。这里将AAPL的数据绘制在最外面，将IBM的数据绘制在里面

也可以用add_subplot来将更多的股票数据放在一起进行对比：

````python
# Figsize for width and height of plot
fig = plt.figure(figsize=(15, 7))

fig.suptitle('Stock price comparison 2007-2017',
             fontsize=20)

ax1 = fig.add_subplot(231)
ax1.set_title('MSFT')

ax1.plot(stock_data['Date'],
         stock_data['MSFT'],
         color='green')

ax2 = fig.add_subplot(232)
ax2.set_title('GOOG')

ax2.plot(stock_data['Date'],
         stock_data['GOOG'],
         color='purple')

ax3 = fig.add_subplot(233)
ax3.set_title('SBUX')

ax3.plot(stock_data['Date'],
         stock_data['SBUX'],
         color='magenta')

ax4 = fig.add_subplot(234)
ax4.set_title('ADBE')

ax4.plot(stock_data['Date'],
         stock_data['ADBE'],
         color='orange')

ax4 = fig.add_subplot(235)
ax4.set_title('NFLX')

ax4.plot(stock_data['Date'],
         stock_data['NFLX'],
         color='chocolate')

ax4 = fig.add_subplot(236)
ax4.set_title('ORCL')

ax4.plot(stock_data['Date'],
         stock_data['ORCL'],
         color='teal',
        )

plt.show()
````

上述代码，每个数据单独绘图，坐标范围也不一样，不利于比较，可以将所有的曲线，画在一个图上，这时候，就需要用不同的颜色(参考[这里](https://matplotlib.org/users/colors.html))，标记（marker,参考[这里](https://matplotlib.org/api/markers_api.html?highlight=marker#module-matplotlib.markers)），以及连接样式（dash style，参考[这里](https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle)）来标记并区分不同的曲线数据。

````python
# Figsize for width and height of plot
fig = plt.figure(figsize=(15, 7))

fig.suptitle('Stock price comparison 2007-2017',
             fontsize=20)


plt.plot(stock_data['Date'], stock_data['MSFT'], 'go-',
         stock_data['Date'], stock_data['GOOG'], 'bv--',
         stock_data['Date'], stock_data['SBUX'], 'm<-.',
         stock_data['Date'], stock_data['ADBE'], 'y>:',
         stock_data['Date'], stock_data['NFLX'], 'c1 ',
         stock_data['Date'], stock_data['ORCL'], 'c2 ',
         )
plt.legend(['MSFT', 'GOOG', 'SBUX', 'ADBE', 'NFLX',  'ORCL', ])

plt.show()
````

## 茎干图/Stem Plot

茎干图能够表示负值，所以很适合用来对差分数据进行展示。还是以游戏数据为例，每年数据的增减可以哦嗯茎干图比较直观的表示出来。




````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

np_data=pd.read_csv('datasets/national_parks.csv')
np_stem= np_data.copy()
````

````python
np_stem[['Badlands', 
         'GrandCanyon',
         'BryceCanyon']] = np_data[['Badlands', 
                                    'GrandCanyon',
                                    'BryceCanyon']].diff()
# diff表示数据差分，用后一个数据减掉前一个数据。

np_stem.head()
````

````python
plt.figure(figsize=(15, 11))

plt.suptitle('Change in Number of Visitors', y=0.94)

plt.subplot(311)
plt.stem(np_stem['Year'],
         np_stem['Badlands'],
         markerfmt='b_',
         linefmt='r--',
         basefmt='g:')
plt.title('GrandCanyon')

plt.subplot(312)
plt.stem(np_stem['Year'],
         np_stem['GrandCanyon'],
         markerfmt='b_',
         linefmt='r--',
         basefmt='g:')
plt.title('Badlands')

plt.subplot(313)
plt.stem(np_stem['Year'],
         np_stem['BryceCanyon'],
         markerfmt='b_',
         linefmt='r--',
         basefmt='g:')
plt.title('BryceCanyon')

plt.show()
````

上图，在横轴上面的部分，表示增长，而在横轴下面的部分则表示下降。

## 散点图/Scatter Plot

散点图比较适合描绘两个变量之间的关系，可以非常直观的描述数据的变化趋势。
#### 使用场景
用于回归类任务的时候，散点图对自变量和因变量的刻画比较直观，同时对离群点的识别也比较方便。

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

countries=pd.read_csv('datasets/countries.csv')
````

````python
countries.head()
````

````python
data_2007=countries[countries['year']==2007]
````

散点图与线图的数据形式类似，不同的是，线图的数据是连续的，在图上用直线连接。

而散点图的数据，一般没有很明显的线可以连接起来。

````python
plt.scatter(data_2007['gdpPerCapita'],data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
````

````python
data_2007['gdpPerCapita'].corr(data_2007['lifeExpectancy'])
````

上图中可以看到，随着gdpPerCapita的增长。lifeExpectancy呈现剧烈的变化趋势，但是有当gdpPerCapita增大到一定值，lifeExpectancy的增长放缓。同时利用corr计算出两个指数的相关系数是0.678。

````python
plt.scatter(np.log10(data_2007['gdpPerCapita']),data_2007['lifeExpectancy'])
plt.title('Life Expectancy vs GDP Per Capita in 2007 ')
plt.xlabel('GDP per Capita($)')
plt.ylabel('Life Expectancy')
plt.show()
````

````python
np.log10(data_2007['gdpPerCapita']).corr(data_2007['lifeExpectancy'])
````

当尝试对gdpPerCapita取log之后，两个指标的数值看上去更接近线性。corr计算的相关系数也支持图中的结论。

上面只取了一年的数据，可以利用3D图像将所有年份的数据都展示出来。

````python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(countries['year'], countries['gdpPerCapita'],
           countries['lifeExpectancy'], s=30)
plt.show()
````

## 柱状图/Bar plot
柱状图用于显示几组数据的分布区别，跟直方图比较容易弄混，事实上，如果数据处理得当的话，柱状图可以在特定情况下代替直方图使用。

### 堆叠柱状图
有时候，光横向对比数据，还不够，还需要对数据进行纵向对比，可以考虑使用堆叠柱状图。


````python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)  # the x locations for the groups
width = 0.35  # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
for idx, rect in enumerate(p1):
    yloc = menMeans[idx] / 2  # 要标注的位置
    label = plt.annotate(menMeans[idx], xy=(idx, yloc))  # 在相应位置标注文字

p2 = plt.bar(ind, womenMeans, width, bottom=menMeans, yerr=womenStd)

for idx, rect in enumerate(p2):
    yloc = menMeans[idx] + womenMeans[idx] / 2
    label = plt.annotate(menMeans[idx], xy=(idx, yloc))

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
````

这样的数据看起来比较直观，但是同时，对于每个柱的数据，看上去比例不是很明确，特别是位置也不明显的情况下，所以有时候需要对数据做一下平均处理。

````python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = np.array((20, 35, 30, 35, 27))
womenMeans = np.array((25, 32, 34, 20, 25))

total = menMeans + womenMeans

menMeans = menMeans / total
womenMeans = womenMeans / total

ind = np.arange(N)  # the x locations for the groups
width = 0.35  # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width)
for idx, rect in enumerate(p1):
    yloc = menMeans[idx] / 2
    label = plt.annotate('{:.1f}%'.format(menMeans[idx] * 100), xy=(idx, yloc))

p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)

for idx, rect in enumerate(p2):
    yloc = menMeans[idx] + womenMeans[idx] / 2
    label = plt.annotate(
        '{:.1f}%'.format(womenMeans[idx] * 100), xy=(idx, yloc))

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 1, 10))  # 注意，这里的坐标范围也需要相应修改
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
````

但是这样处理的数据，纵向比较方便看比例，横向的可比性就没有了。

## Grouped bar chart/分组柱状图

上面的柱状图仍然不够直观，我们想很直观的看两个数据的对比，这个时候，可以尝试能够直接将数据画在一起做对比的分组柱状图。

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width,
                label='Men', edgecolor='red', linewidth=2)
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
````

从上面几个图可以看出，柱状图，就是直接在指定的位置画一个矩形，同时返回代表这个矩形的各个对象。可以通过各个对象获取矩形的参数来对矩形做进一步的设置。

## Histogram/直方图

histgrom用于将一系列的数据分组计数，非常适合用来统计一个数据序列中每个区间的数据数量。
### 使用场景
用于需要统计一个序列数据中，各个区间的数据个数的场景。

下面以某个游戏的销量数据为例。


````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

np_data=pd.read_csv('datasets/national_parks.csv')
````

````python
np_data.head()
````

````python
plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10)
plt.show()
````

上图中可以看出，销量在200w、250w和400w左右的年份比较多，都在8～10个年份，450w左右的年份最多有高达16个年份。同时销量高于500w以及在300w，350w的年份较少，2个年份或者更少。

### histgrom的组成
histogram在绘制的时候，同时还会返回图像中的一些组成部分：

* __n:__ 每个区间有多少个数据
* __bins:__ 每个区间的中间点
* __patches:__ historgam里面的每个柱子。


````python
n, bins, patches = plt.hist(np_data['GrandCanyon'], 
                            facecolor='peru', 
                            edgecolor='blue', 
                            bins=10)

print('n: ', n)
print('bins: ', bins)
print('patches: ', patches)
````

histgrom不止可以计算数据在各个区间的计数，还可以使用cumulative参数计算累计计数。从累计基数，我们可以更好的理解数据的增长趋势。

````python
plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10,
         cumulative=True)
plt.show()
````

有时我们需要了解某个范围内的数据，那么在绘制histogrom的时候，我们需要指定数据的范围。

````python
plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10,
         range=(2000000, 5000000))

plt.show()
````

在有多个数据的情况下，我们经常需要对比多个数据，histogram支持多个数据同时显示，但是需要注意，同时显示两个数据的时候，我们需要设置数据透明度，避免遮挡。

````python
plt.hist(np_data['BryceCanyon'], 
         facecolor='orangered', 
         edgecolor='maroon', 
         bins=8)

plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10, 
         alpha = 0.3)  # 透明度

plt.show()
````

注意上面代码里面的alpha，因为GrandCanyon的histogram是后画的，不指定透明度，则GrandCanyon的histogram会挡住BryceCanyon的图像。

除了一维的histogram，matplotlib还可以绘制二维的histogram。
下图中的线上的矩形框的宽，高，分别代表落入纵坐标和横坐标的数据的个数。

````python
plt.hist2d(np_data['Year'], np_data['GrandCanyon'], bins=(50, 50))
plt.plot(np_data['Year'], np_data['GrandCanyon'])
plt.legend()
plt.show()
````

## Pie Chart

饼图可以非常直观的展示数据分类之间的比例关系。

下面的例子中，这个图像展示了每个分类的比例以及及分布关系：


````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov= pd.read_csv('datasets/types_movies.csv')
````

````python
t_mov
````

````python
plt.figure()
plt.pie(t_mov['Percentage'],
        labels=t_mov['Sector'])

plt.axis('equal')
plt.show()
````

### pie chart的组成
饼图在绘制的时候，还会返回一些对象，代表饼图的组成。


- __wedges:__ 饼图中，各个饼的对象列表
- __texts:__ 各个饼的标签
- __autotexts:__ 各个饼的数值，需要注意这一项只有饼图绘制参数中autopct不为None的时候才有效

````python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.2f')

plt.axis('equal')
````

````python
print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
````

### piechart的自定义
pie chart中，colors可以用来指定每个pie的颜色，而autopct选项可以用来指定每个pie的百分比的显示格式。
> 如果不是为了统一颜色样式的话，不建议自己指定颜色，matplotlib自动提供的颜色保证了每个pie都很容易区分。

````python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered', 'chocolate']

plt.pie(t_mov['Percentage'],
        labels=t_mov['Sector'],
        colors=colors,
        autopct='%.2f')

plt.axis('equal')

plt.show()
````

### explode参数，pie chart的突出显示

我们可以使用explode这个参数，指定每个pie离中心的距离，用于突出显示某个pie

````python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
````

## 箱图（box plot）

箱图用来理解数据的分布状况。

### 使用场景
当需要获取数据的分布信息以及检查离群点/异常值分布的时候。

箱图的解释

![](1_fCE_5juz235c6cmaOP_PDQ.png)


箱图用五个典型数值点勾勒数据的分布属性：
- 最大观测值(Maximum)
- 上四分位数(Lower Quartile/Q3)，有25%的数据大于这个值
- 中位数(Median)，有50%的数据小于这个值
- 下四分位数(Lower Quartile/Q1)，有25%的数据小于这个值
- 最小观测值(Minimum)

位于Minimum以下或Maximum以上的部分称为离群点（Outlier），为数据中的异常值。

Q1/Q3中间的部分，被称为中间四分位数极差（ Inter Quartile Range/IQR）。Q1到Minimum以及Q3到Maximum的部分，被称为胡须(whisker)。

需要注意的是，最大最小观测值是计算出来的数值，规则如下：
- Minimum = Q1 - 1.5*IQR，如果存在Outlier小于Minimum，则胡须下限为Minimum，Outlier单独以点汇出。如果没有比Minimum小的数，则胡须下限为Minimum。
- Maximum = Q3 -1.5*IQR，如果存在Outlier大于Maximum，则胡须上限为Maximum，Outlier单独以点汇出。如果没有比Maximum大的数，则胡须上限为Maximum。

## Violin plot/提琴图
提琴图因为纺槌状形似提琴而得名，其宽度代表了相应位置的数据分布，因此比箱图能看到更多的细节信息。

#### 使用场景
箱图不足以详细理解数据的时候，提琴图作为箱图的补充。

````python
vp = plt.violinplot(exam_scores_array,
                    showmedians=True)

plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])

plt.show()    
````

可以根据横向的宽度判断数据在某个数值附近的密度，同时也可以直观的看到更详细的数据分布。
上图中，阅读的分数分布比较平均，数学分数则呈现比较明显的扎堆的情形。


有时候，我们也把这两种图画在一起

````python
colors = ['blue', 'grey', 'lawngreen']

bp = plt.boxplot(exam_scores_array,
                 patch_artist=True,
                 notch=True,
                 showmeans=True)

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=colors[i])
    bp['caps'][2*i + 1].set(color=colors[i])
plt.xticks([1, 2, 3], ['Math', 'Reading', 'Writing'])

vp = plt.violinplot(exam_scores_array,
                    showmedians=True)

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i])
plt.show()
````

## stack plot/堆叠图

堆叠图以堆叠的形式展示多个时间序列的联合分布信息。

这里以游戏数据的年度销量为例，分别展示了每种游戏的销量及其大小比例信息。


````python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np_data= pd.read_csv('datasets/national_parks.csv')
````

````python
np_data.head()
````

````python
x = np_data['Year']
````

````python
# 堆叠图要求的输入多多维数据，这里用vstack来实现
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])
````

````python
# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc='upper left')

plt.show()
````

由于堆叠图，表示的数据之间的面积，上下边界的起点是其他数据的边界，而不是X轴，所以看起来不是很直观。

这里可以做一个百分比处理，将图像面积表示为百分比，数据可以更加直观，如下图，可以很明显的看出各年份，各游戏的销量占比。

>但是这样只能横向对比各数据之间的比例，同一个数据之间的数值变化就没有意义了。

````python
plt.figure(figsize=(10,7))

data_perc = np_data.divide(np_data.sum(axis=1), axis=0)
# 注意，这里百分比的处理是在各组数据之间做的，同组的数据已经不再有可比性

plt.stackplot(x,
              data_perc["Badlands"],data_perc["GrandCanyon"],data_perc["BryceCanyon"],
              edgecolor='black',
              colors=colors,
              labels=labels)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
````

## 次坐标/TwinAxis


## 使用场景
当我们需要观察比较两个有同样的x轴，不同的y轴数据的时候。

例如人口和GDP数据，其横轴都是年份。

这里一气象数据为例，横轴为日期，纵轴分别显示温度和风速。


````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

austin_weather = pd.read_csv('datasets/austin_weather.csv')
````

````python
austin_weather.head()
````

Extracting Date,Avg Temperature and Avg Wind Speed columns

````python
austin_weather = austin_weather[['Date', 'TempAvgF', 'WindAvgMPH']].head(30)
austin_weather
````

Plotting two lines with 2 separate Y axes with the same X axis
First create one axis and then use the twinx() function to create another axis sharing the same X axis

````python
# Subplot for Plotting Figure
fig, ax_tempF = plt.subplots()

# Similar to fig=plt.figure(figsize=(12,6))
fig.set_figwidth(12)
fig.set_figheight(6)

# set x label which is common
ax_tempF.set_xlabel('Date')

# bottom= false disables ticks and labelbottom disables x-axis labels
ax_tempF.tick_params(axis='x',
                     bottom=False,
                     labelbottom=False)

# set left y-axis label
ax_tempF.set_ylabel('Temp (F)',
                    color='red',
                    size='x-large')

# set labelcolor and labelsize to the left Y-axis
ax_tempF.tick_params(axis='y',
                     labelcolor='red',
                     labelsize='large')

# plot AvgTemp on Y-axis to the left
ax_tempF.plot(austin_weather['Date'],
              austin_weather['TempAvgF'],
              color='red')

# twinx sets the same x-axis for both plots
ax_precip = ax_tempF.twinx()

# set Right y-axis label
ax_precip.set_ylabel('Avg Wind Speed (MPH)',
                     color='blue',
                     size='x-large')

# set labelcolor and labelsize to the Right Y-axis
ax_precip.tick_params(axis='y',
                      labelcolor='blue',
                      labelsize='large')

# plot WindAVg on Y-axis to the Right
ax_precip.plot(austin_weather['Date'],
               austin_weather['WindAvgMPH'],
               color='blue')

fig.legend(bbox_to_anchor=(1, 1), bbox_transform=ax_tempF.transAxes)

plt.show()
````

从上图可以看出，风速和数据有明显的负相关，风速最大的时候，温度最低，反之亦然。

次坐标不仅可以用于显示两个不同的数据，还可以用于显示同一个数据的不同度量。

如一个观测值有两种不同的计量单位，则可以同时显示出来。

上图显示的温度为华氏度，下面将其转换为摄氏度显示。

首先定义一个转换函数。

````python
def fahrenheit2celsius(f):
    return (f - 32) * 5 / 9
````

````python
# Subplot for Plotting Figure
fig, ax_tempF = plt.subplots()

# Similar to fig=plt.figure(figsize=(12,6))
fig.set_figwidth(12)
fig.set_figheight(6)

# set x label which is common
ax_tempF.set_xlabel('Date')

# bottom= false disables ticks and labelbottom disables x-axis labels
ax_tempF.tick_params(axis='x',
                     bottom=False,
                     labelbottom=False)

# set left y-axis label
ax_tempF.set_ylabel('Temp (F)',
                    color='red',
                    size='x-large')

# set labelcolor and labelsize to the left Y-axis
ax_tempF.tick_params(axis='y',
                     labelcolor='red',
                     labelsize='large')

# plot AvgTemp on Y-axis to the left
ax_tempF.plot(austin_weather['Date'],
              austin_weather['TempAvgF'],
              color='green')

# twinx sets the same x-axis for both plots
ax_tempC = ax_tempF.twinx()

# get ymin and ymin of AvgTemp in Fahrenheit
ymin, ymax = ax_tempF.get_ylim()

# set ymin and ymin of AvgTemp in Fahrenheit to AvgTemp in Celsius
ax_tempC.set_ylim((fahrenheit2celsius(ymin),
                   fahrenheit2celsius(ymax)))

# set Right y-axis label
ax_tempC.set_ylabel('Temp (C)',
                    color='blue',
                    size='x-large')

# plot AvgTemp in Celsius on Y-axis to the Right
ax_tempC.tick_params(axis='y',
                     labelcolor='blue',
                     labelsize='large')

plt.show()
````

````python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import time

def generate(X, Y, phi):
    R = 1 - np.sqrt(X**2 + Y**2)
    return np.cos(2 * np.pi * X + phi) * R

plt.ion()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

xs = np.linspace(-1, 1, 50)
ys = np.linspace(-1, 1, 50)
X, Y = np.meshgrid(xs, ys)
Z = generate(X, Y, 0.0)

wframe = None
tstart = time.time()
for phi in np.linspace(0, 360 / 2 / np.pi, 100):

    oldcol = wframe

    Z = generate(X, Y, phi)
    wframe = ax.plot_wireframe(X, Y, Z, rstride=2, cstride=2)

    # Remove old line collection before drawing
    if oldcol is not None:
        ax.collections.remove(oldcol)

    plt.draw()

print ('FPS: %f' % (100 / (time.time() - tstart)))
````
